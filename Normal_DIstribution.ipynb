```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
def generateNormalDist(sampleSize):
    x=0 
    i=0 
    while i < sampleSize: 
        u = random.uniform(-6,6) 
        x=x+u 
        i=i+1
       # print(i)
    return (x)/sampleSize

numOfSimulations =0 
normDist=[] 
while numOfSimulations <100 :
    normDist.append(generateNormalDist(12))
    numOfSimulations+=1

sb.distplot(normDist)
#plt.hist(normDist)
```

```python
#Box muller for standard normal
i=0
X1=[]
X2=[]
while i < 1000:
    u1 = random.uniform(0,1)
    u2 = random.uniform(0,1)
    X1.append(np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2))
    X2.append(np.sqrt(-2*np.log(u1))*np.sin(2*np.pi*u2))
    i=i+1

sb.distplot(X1)
sb.distplot(X2)



```
