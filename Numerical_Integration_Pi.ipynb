```python
import numpy as np
from numpy import random
%matplotlib inline
import matplotlib.pyplot as plt
import time
```

```python
start_time=time.time()

```

```python
def Number_integration(NumberOfSamples):
    def func(x):
        return np.sqrt(1-x**2)
    i=0
    integralValue=0;
    while i<NumberOfSamples:
        integralValue+= func(random.uniform(0,1))
        i+=1
    return integralValue*4/float(NumberOfSamples)
```

```python
count=0
IntegralValues=[]
while count<1000:
    IntegralValues.append(Number_integration(10000))
    count+=1

meanPi = np.mean(IntegralValues)
plt.figure(figsize=(20,12))
plt.hlines(meanPi,color='red',xmin=0,xmax=1000)
plt.plot(IntegralValues,'bo')
```
