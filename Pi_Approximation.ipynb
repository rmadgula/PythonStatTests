```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import time
```

```python
start_time=time.time()
def approximate_pi(NumOfIterations):
    circleDotsX = []
    circleDotsY=[]
    squareDotsX=[]
    squareDotsY=[]
    i=0

    while i < NumOfIterations:
        randNumX = random.uniform(-1,1)
        randNumY = random.uniform(-1,1)
    
        if (randNumX**2 + randNumY**2)<=1:
            circleDotsX.append(randNumX) 
            circleDotsY.append(randNumY)
        else:
            squareDotsX.append(randNumX)
            squareDotsY.append(randNumY)

        i=i+1
 
    #plt.plot(circleDotsX,circleDotsY,'b.')
    #plt.plot(squareDotsX,squareDotsY,'g.')
    #plt.grid()
    return 4*len(circleDotsX)/float(NumOfIterations)

```

```python
ValueOfPi=[]
piCount=0
while piCount<1000:
    ValueOfPi.append(approximate_pi(10000))
    piCount=piCount+1

meanPI = np.mean(ValueOfPi)
```

```python
plt.figure(figsize=(20,10))
plt.plot(ValueOfPi,'bo')

plt.hlines(y=meanPI,xmin=0,xmax=1000,colors='r',linestyle='solid')



```
