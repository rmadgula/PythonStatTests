```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
def generateNormalDist(sampleSize):
    x=0 
    i=0 
    while i < sampleSize: 
        u = random.uniform(-6,6) 
        x=x+u 
        i=i+1
       # print(i)
    return (x)/sampleSize

numOfSimulations =0 
logNormDist=[] 
while numOfSimulations <100 :
    logNormDist.append(np.exp(generateNormalDist(12)))
    numOfSimulations+=1

sb.distplot(logNormDist)
#plt.hist(normDist)
```
