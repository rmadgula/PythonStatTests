```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sps
%matplotlib inline
```

```python
def exponenGenerator(size,lambdaV):
    i=0
    x=[]
    while i<size:
        x1 = -(1/lambdaV)*np.log(random.uniform(0,1))
        x.append(x1)
        i+=1
    return x


y=exponenGenerator(100,2.8)
sns.distplot(y)
```
