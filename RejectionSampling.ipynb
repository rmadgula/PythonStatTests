```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import random
```

```python
def fxBICGX(x):
    return (4-np.sin(20*x))/5

acceptedz=[]
rejectedz=[]
sampleCount=0
while len(acceptedz)<500:
   
    u =  random.uniform(0,1)
    u1 = random.uniform(0,1)
    u2 = random.uniform(0,1)
    #simulate standard normal for t5 education
    z1 = np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2)
    #z2 = np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2)

    if u<=fxBICGX(z1):
        print('accept z')
        acceptedz.append(z1)
    else:
        print('reject z')
        rejectedz.append(z1)
    
    sampleCount= sampleCount+1
    
    
```
