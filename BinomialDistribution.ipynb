```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import random
```

```python
#Trying to simulate 100 Bernoulli(p) samples, here p=0.24 , probability of success
def BernoulliDist(p):
    u = random.uniform(0,1)
    if u <= p:
        x  = 1 
    else:
        x =0
    return x

#Trying to simulate 100 Bernoulli(p) samples, here p=0.24 ,n=10 -out of 10 trials we want probability of success to be 0.24
def BinomialDist(n,p):
    countOfTrials=0
    binomVar=0
    while countOfTrials<n:
        binomVar+=  BernoulliDist(p)
        countOfTrials = countOfTrials+1
    return binomVar

numSamples=100
SampleOut=[]
while numSamples>0:
    SampleOut.append(BinomialDist(10,0.24))
    numSamples=numSamples-1



sns.distplot(SampleOut,kde=True)





```
