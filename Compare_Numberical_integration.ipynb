```python
import numpy as np
from numpy import random
%matplotlib inline
import matplotlib.pyplot as plt
import time
```

```python
start_time=time.time()

```

```python
def Number_integration_MC(NumberOfSamples):
    def func(x):
        return (x-x**4)* np.power(np.exp,-(x*x)/2)
    i=0
    integralValue=0;
    while i<NumberOfSamples:
        integralValue+= func(random.uniform(-1,1))
        i+=1
    return integralValue
```

```python
count=0
IntegralValues10=[]
IntegralValues100=[]
IntegralValues1000=[]
while count<500:
    IntegralValues10.append(Number_integration_MC(10))
    IntegralValues100.append(Number_integration_MC(100))
    IntegralValues1000.append(Number_integration_MC(1000))
    count+=1
plt.plot(IntegralValues10)
plt.plot(IntegralValues100)
plt.plot(IntegralValues1000)
```
